```python
import pandas as pd
```

```python
from itertools import product, combinations

original_positions_players_map = {
    "A":["W", "C", "H", "Lé", "G", "T", "M"],
    "P":["W", "C", "Lo", "G", "M"],
    "O":["W", "C", "Lé", "T", "M"],
    "L":["Lo", "G", "T"],
    "C":["H", "Lé", "T"]
}

positions_players_map = {
    "A":["W", "C", "Lé", "G", "T", "M"],
    "P":["W", "C", "Lo", "G", "M"],
    "O":["W", "C", "Lé", "T", "M"],
    "L":["Lo", "G", "T"],
    "C":["Lé", "T", "D", "B"]
}

positions_players_scores = {
    "A":[1, 2, 1, 3, 2, 1],
    "P":[2, 3, 1, 1, 3],
    "O":[3, 1, 2, 1, 2],
    "L":[2, 2, 3],
    "C":[3, 4, 1, 1]
}
```

```python
all_combinations = list(product(combinations(positions_players_map["A"], r=2), positions_players_map["P"], positions_players_map["O"], positions_players_map["L"], combinations(positions_players_map["C"], r=2)))

all_scores = list(product(combinations(positions_players_scores["A"], r=2), positions_players_scores["P"], positions_players_scores["O"], positions_players_scores["L"], combinations(positions_players_scores["C"], r=2)))

all_combinations, all_scores
```

```python
all_scores = list(map(lambda score: sum(score[0]) + sum(score[1:-1]) + sum(score[-1]), all_scores))
all_combinations = list(map(lambda comb : (comb[0][0], comb[0][1], comb[1], comb[2], comb[3], comb[4][0], comb[4][1]), all_combinations))
```

```python
all_combinations_with_scores = list(zip(all_combinations, all_scores))
```

```python
from collections import Counter

def no_clones(comb:tuple)->bool:
    return len(Counter(comb).values()) == 7
```

```python
possible_combinations = [(comb, score) for comb, score in all_combinations_with_scores if no_clones(comb)]
```
